```python
#But pos_tags and lemmatizers are different, so can't be clubbed together
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
    
#But a little problem with the above function
w = "better"
pos_tag(w) #so you need to pass an array, to apply the operation on complete word
```
