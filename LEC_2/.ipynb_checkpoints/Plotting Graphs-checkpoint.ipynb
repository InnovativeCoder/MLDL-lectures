```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y) #a scatter plot only of dats, no line
plt.show() #to show the graphs after all the customizations
```

```python
x = [1,2,3]
y = [2,4,6]

plt.plot(x,y) # plot a line
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]

x2 = [3,4,5]
y2 = [2,4,6]

plt.plot(x,y,"r+") # a dash dot line
plt.plot(x2,y2,"b--") # a dashed line
plt.show()
```

```python
#plot function x^3
import numpy as np
x = np.array([1,2,3,4])
y = x ** 3
print(y)
plt.plot(x, y)
plt.show()
```

```python
#Exactly like cube
x = np.arange(0,5,0.1)
y = x ** 3
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
# color, marker, linewidth, labels for axis, datasets and the title, axis, grid, text
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y  = x **3
y2 = x**2
#plt.plot(x, y, color = "black", marker="p")
plt.plot(x, y, color = "black", linewidth = 5, label = "x^3")

plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")

#plt.axis([0, 10, 0, 200])

plt.ylabel("y")
plt.xlabel("x")

plt.grid()

plt.text(2, 80, "Text", fontsize=12)

plt.legend()
plt.title("Matplotlib demo")
plt.show()
```

```python
# pie graphs
sizes = [3, 4, 6, 2]
plt.pie(sizes)
plt.show()
```

```python
# customized pie graphs
labels = ["A","B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["blue", "purple", "red", "pink"]
explode= [0.1, 0.2, 0, 0]
plt.title("Split among classes")
plt.pie(sizes, colors=colors, explode = explode, labels=labels, autopct="%.2f%%",  startangle=100)
plt.axis("equal")
plt.show()
```
