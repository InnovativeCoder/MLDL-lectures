```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(max_iter=1000) #change max_iter till it converges
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
#to check all the coeffoecients
clf.coefs_
```

```python
len(clf.coefs_) # there are 2 layers, 1 hidden layer
```

```python
clf.coefs_[0].shape, clf.coefs_[1].shape #4 input features , 100 hidden units, and 3 outputs
```

```python
#looking for biases 
clf.intercepts_[0]
```

```python
clf.intercepts_[0].shape, clf.intercepts_[1].shape #one bias going to every unit
```
